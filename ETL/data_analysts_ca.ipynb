```python
import pandas as pd
```

#### Data Cleaning

```python
df = pd.read_csv("DataAnalyst.csv")
```

```python
df["Location"].unique()
```

```python
new = df["Location"].str.split(",", n = 1, expand = True) 
```

```python
df["City"] = new[0] 
df["State"] = new[1]
```

```python
df_CA = df.loc[df["State"] == " CA"]
```

```python
df_CA.head()
```

```python
salary = df_CA["Salary Estimate"].str.split("-", n = 1, expand = True) 
```

```python
df_CA["SalaryEstimate_low"] = salary[0]
df_CA["SalaryEstimate_high"] = salary[1]
a = df_CA["SalaryEstimate_low"].str.split("K", n = 1, expand = True) 
b = df_CA["SalaryEstimate_high"].str.split("K", n = 1, expand = True) 
df_CA["SalaryEstimate_low"] = a[0]
df_CA["SalaryEstimate_high"] = b[0]
c = df_CA["SalaryEstimate_low"].str.split("$", n = 1, expand = True) 
d = df_CA["SalaryEstimate_high"].str.split("$", n = 1, expand = True) 
df_CA["SalaryEstimate_low"] = c[1]
df_CA["SalaryEstimate_high"] = d[1]
df_CA[["SalaryEstimate_low", "SalaryEstimate_high"]] = df_CA[["SalaryEstimate_low", "SalaryEstimate_high"]].apply(pd.to_numeric, errors='coerce')
df_CA[["SalaryEstimate_low","SalaryEstimate_high"]] = df_CA[["SalaryEstimate_low","SalaryEstimate_high"]]*1000
```

```python
df_CA["Revenue"] = df_CA["Revenue"].str.split("billion", n = 1, expand = True) 
revenue = df_CA["Revenue"].str.split("to", n = 1, expand = True) 
df_CA["Revenue_low"] = revenue[0]
df_CA["Revenue_high"] = revenue[1]
x = df_CA["Revenue_high"].str.split(" ", n = 1, expand = True) 
df_CA["Revenue_high"] = x[1]
y = df_CA["Revenue_high"].str.split(" ", n = 1, expand = True)
df_CA["Revenue_high"] = y[0]
z = df_CA["Revenue_low"].str.split("$", n = 1, expand = True) 
t = df_CA["Revenue_high"].str.split("$", n = 1, expand = True) 
df_CA["Revenue_low"] = z[1]
df_CA["Revenue_high"] = t[1]
df_CA[["Revenue_low", "Revenue_high"]] = df_CA[["Revenue_low", "Revenue_high"]].apply(pd.to_numeric, errors='coerce')
df_CA[["Revenue_low","Revenue_high"]] = df_CA[["Revenue_low","Revenue_high"]]*1000000
```

```python
df_CA.head()
```

```python
df_CA = df_CA.drop(columns=["Revenue","Salary Estimate"])
```
