```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import pymongo 
import requests
from config import api_key
from selenium import webdriver
import datetime as dt
from datetime import datetime,timedelta
import numpy as np
```

```python
executable_path = {'executable_path': '/Users/moo/Desktop/final_project/chromedriver'}
browser = Browser('chrome', **executable_path, headless=True)
```

```python
#url='https://www.indeed.com/jobs?q=data%20analyst&l=CA&radius=150'
url= 'https://www.indeed.com/jobs?q=data%20analyst&l=Sacramento%2C%20CA&radius=50'
browser.visit(url)
```

```python
html = browser.html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
titles = soup.find_all("h2", class_ = "title")
print(titles)
```

```python
for title in titles:
    job_title = title.get_text()
    print(job_title)
```

```python
#TITLE

job_titles = [0]
 
 # gether the job results /// from 1-10 pages  
for x in range(1, 201):
    titles = soup.find_all("h2", class_ = "title")
    for title in titles:
        job_titles.append(title.text)
        print('page:', x, '-------------')
        print(title.text)
        browser.find_by_xpath('/html/body/table[2]/tbody/tr/td/table/tbody/tr/td[1]/nav/div/ul/li[6]/a/span/span')
```

```python
job_posted = []


#loop through the returned result
for x in range(1, 201):
    locations = soup.find_all("span", class_ = "location accessible-contrast-color-location")
    for location in locations:
        job_posted.append(location.text)
        print('page:', x, '-------------')
        print(location.text)
        browser.find_by_xpath('/html/body/table[2]/tbody/tr/td/table/tbody/tr/td[1]/div[7]/div[1]/span[1]')
```

```python
job_company = []


#loop through the returned result
for x in range(1, 201):
    company = soup.find_all("span", class_ = "company")
    for c in company:
        job_company.append(c.text)
        print('page:', x, '-------------')
        print(c.text)
        browser.find_by_xpath('/html/body/table[2]/tbody/tr/td/table/tbody/tr/td[1]/div[7]/div[1]/span[1]')
```

```python
job_salary = []


#loop through the returned result
for x in range(1, 201):
    salary = soup.find_all("span", class_ = "salary")
    for s in salary:
        job_salary.append(s.text)
        print('page:', x, '-------------')
        print(s.text)
        browser.find_by_xpath('/html/body/table[2]/tbody/tr/td/table/tbody/tr/td[1]/div[7]/div[1]/span[1]')
```

```python
clean_job_titles = []

#loop through result
for job in job_titles[1:]:
    clean_job_title = job.replace('\n', '')
    clean_job_titles.append(clean_job_title)
```

```python
clean_job_titles
```

```python
clean_location = []

#loop through result
for lo in job_posted[1:]:
    clean_l= lo.replace('\n', '')
    clean_location.append(clean_l)
print(clean_location)
```

```python
clean_company = []

#loop through result
for company in job_company[1:]:
    clean_com= company.replace('\n', '')
    clean_company.append(clean_com)
print(clean_company)
```

```python
clean_salary = []

#loop through result
for ss in job_salary[1:]:
    clean_s= ss.replace('\n', '')
    clean_salary.append(clean_s)
print(clean_salary)
```

```python
new_list = list(zip(clean_job_titles,clean_location,clean_company,clean_salary))
new_list
```

```python
job_df = pd.DataFrame(new_list, columns=['job_title', 'location','company','salary_range'])
job_df
```

```python
job_df.to_csv('sac_job_scraping.csv',index=False)
```
