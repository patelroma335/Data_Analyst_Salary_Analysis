```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#df = pd.read_csv("indeed_job_dataset.csv")

# Read CSV
indeed_data = pd.read_csv("indeed_CA.csv")
CA_data = pd.read_csv("df_CA.csv")
```

```python
indeed_data.head()
```

```python
# Drop all rows with missing information
indeed_data = indeed_data.dropna(how='any')
```

```python
# Verify dropped rows
indeed_data.count()
```

```python
plt.rcParams["figure.figsize"] = (20,10)
indeed_data['Company'].value_counts().sort_values(ascending=False).head(25).plot.bar(facecolor="purple", alpha=0.50, align="center")

plt.title('Companies with the Most Jobs Offered', size=20)
plt.xlabel("Company Name")
plt.ylabel("Number of Most Jobs Offered")

plt.show()
plt.savefig('Companies with Most jobs Offered.png')

indeed_data['Company'].value_counts().sort_values(ascending=False).head(5)
```

## What are the top 5 most common jobs?

```python
CA_data['Job Title'].value_counts().head(5)
```

```python
Job_type = ["Data Analyst", "Senior Data Analyst", "Sr. Data Analyst", "Business Data Analyst", "Junior Data Analyst"]
Numbers = [121, 28, 12, 10, 6]
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue", "orange"]
explode = (0.05, 0, 0, 0, 0)
```

```python
plt.title("Top 5 Most Common Jobs", size=20)
plt.pie(Numbers, explode=explode, labels=Job_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.legend(Job_type, loc='best', fontsize=12)
plt.show()
plt.savefig('Top 5 Most Common Jobs.png')
```

## What is the Min and Max Salary Distribution?

```python
plt.rcParams["figure.figsize"] = (10,8)
plt.boxplot([CA_data['SalaryEstimate_low'],CA_data['SalaryEstimate_high']], labels=['Min. Salary', 'Max. Salary'])

plt.title('Salary Distribution', size=20)
plt.show()
plt.savefig('Salary Distribution.png')
```

```python
plt.rcParams["figure.figsize"] = (20,6)

CA_data.boxplot(column='SalaryEstimate_low', by='Industry')

plt.title('Lower Salary Estimate by Industry', size=20)
plt.xlabel('Industry')
plt.ylabel('Lower Salary Estimate')

plt.xticks(rotation='vertical')
plt.show()
plt.savefig('Lower Salary Estimate by Industry.png')
```

```python
plt.rcParams["figure.figsize"] = (20,6)

CA_data.boxplot(column='SalaryEstimate_high', by='Industry')

plt.title('Higher Salary Estimate by Industry', size=20)
plt.xlabel('Industry')
plt.ylabel('Higher Salary Estimate')

plt.xticks(rotation='vertical')
plt.show()
plt.savefig('Higher Salary Estimate by Industry.png')
```

```python
lowest_industry = CA_data.groupby('Industry')['SalaryEstimate_low'].mean().idxmax()
lowest_ind_val = CA_data.groupby('Industry')['SalaryEstimate_low'].mean().max()
print('The lowest expectedly paid industry is {} with ${}'.format(lowest_industry, lowest_ind_val))
```

```python
highest_industry = CA_data.groupby('Industry')['SalaryEstimate_high'].mean().idxmax()
highest_ind_val = CA_data.groupby('Industry')['SalaryEstimate_high'].mean().max()
print('The highest expectedly paid industry is {} with ${}'.format(highest_industry, highest_ind_val))
```

## Which Industry has the Best Rating?

```python
plt.rcParams["figure.figsize"] = (20,5)

CA_data.boxplot(column='Rating', by='Industry')

plt.title('Ratings by Industry')
plt.xlabel('Industry')
plt.ylabel('Ratings')

plt.xticks(rotation='vertical')
plt.show()
plt.savefig('Rating by Industry.png')
```

   ## Pay Difference by location

```python
plt.rcParams["figure.figsize"] = (20,7)

CA_data.groupby('City')['SalaryEstimate_low'].mean().plot.bar(facecolor="orange", alpha=0.75, align="center")

plt.title('Min.Expected Salary by Cities in CA', size=20)
plt.xlabel('Cities in CA')
plt.ylabel('USD')
plt.show()
plt.savefig('Min.Expected Salary by Cities in CA.png')
```

```python
plt.rcParams["figure.figsize"] = (25,6)

CA_data.groupby('City')['SalaryEstimate_high'].mean().plot.bar()

plt.title('Max.Expected Salary by Cities in CA', size=20)
plt.xlabel('Cities in CA')
plt.ylabel('USD')
plt.show()
plt.savefig('Max.Expected Salary by Cities in CA.png')
```
