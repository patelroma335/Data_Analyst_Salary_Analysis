```python
import io
df = pd.read_csv(io.BytesIO(uploaded['dummy_data.csv']))
```

```python
df.head()
```

```python
# Pre-Processing
```

```python
labels=["Job_Type","Job_Title","Company"]
X = df.drop(labels=labels, axis=1)
y = df["Job_Type"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Label-Encode the dataset
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Define X and y Features
```

```python
X_features = X_train_scaled.shape[1]
X_features 
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
y_features = y_train_categorical.shape[1]
y_features
```

```python
# Create model and add layers
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=10, activation='relu', input_dim=X_features))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=y_features, activation='softmax'))
```

```python
# Compile and fit the model
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=20,
    shuffle=True
)
```

```python
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
```

```python
# Quantify our Trained Model
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Make Predictions
```

```python
predictions = model.predict_classes(X_test_scaled)
```
